```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel(r"C:\Users\Aaditya\Desktop\Notes\apvendordb_threshold_output.xlsx")
dataset.head()
```

```python
final_dataset = dataset.sort_values(by=['vendodb_threshold'], ascending=True)
final_dataset.head(10)
```

```python
thresold = final_dataset.loc[:,["vendodb_threshold"]].values
my_accuracy = final_dataset.loc[:,["accuracy"]].values
my_coverage = final_dataset.loc[:,["coverage"]].values
my_overall_accuracy = final_dataset.loc[:,["overall_accuracy"]].values
```

```python
plt.plot(thresold, my_accuracy, label='accuracy')
plt.plot(thresold, my_coverage, label='coverage')
plt.plot(thresold, my_overall_accuracy, label='overall accuracy')
plt.xticks(range(3, 18))
plt.yticks(range(0, 75, 5))
plt.legend()
plt.grid()
plt.savefig('product.png')
plt.show()
```
