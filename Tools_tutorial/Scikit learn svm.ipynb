```python
#SVC (support vector classifier) from SVM(support vector Machine)
from sklearn import svm
import matplotlib.pyplot as plt
x = [1, 5, 1.5, 8, 1, 9]
y = [2, 8, 1.8, 8, 0.6, 11]
plt.scatter(x,y)
plt.show()
```

```python
import numpy as np
X = np.array([[1,2],
             [5,8],
             [1.5,1.8],
             [8,8],
             [1,0.6],
             [9,11]])
y = [0,1,0,1,0,1]
```

```python
clf = svm.SVC(kernel = 'linear', C = 1.0)
clf.fit(X,y)

clf.predict([[0.10,0.10]])
```

```python
from mpl_toolkits import mplot3d
import numpy as np
fig = plt.figure()
ax = plt.axes(projection = '3d')
z = np.linspace(10,25,1000)
y = np.cos(z)
x = np.sin(z)
ax.plot3D(z,y,x)

z1 = np.linspace(10,25,10)
y1 = np.cos(z1)
x1 = np.sin(z1)
ax.plot3D(z1,y1,x1,'*')

plt.show()
```
