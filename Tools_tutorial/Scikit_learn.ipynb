```python
import pandas as pd
import matplotlib.pyplot as plt


from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

iris = pd.read_csv("/home/aaditya/Ml/data/Iris.csv")
iris.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tmp = iris.drop('Id', axis=1)
sns.pairplot(tmp, hue='Species')
plt.show()
```

```python
sns.violinplot(x = 'SepalLengthCm',y = 'Species', data=iris, inner = 'quartile')
plt.show()
sns.violinplot(x = 'SepalWidthCm',y = 'Species', data=iris, inner = 'quartile')
plt.show()
sns.violinplot(x = 'PetalLengthCm',y = 'Species', data=iris, inner = 'quartile')
plt.show()
sns.violinplot(x = 'PetalWidthCm', y='Species', data=iris, inner='quartile')
plt.show()
```

```python
X = iris.drop(['Id','Species'], axis=1)
y = iris['Species']
print(X.shape)
print(y.shape)
```

```python
k_range = list(range(1,26))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X,y)
    y_pred = knn.predict(X)
    scores.append(metrics.accuracy_score(y,y_pred))

plt.plot(k_range, scores)
plt.xlabel('K_range')
plt.ylabel('scores')
plt.title('accutracy scores')
plt.show()
```

```python
logreg = LogisticRegression()
logreg.fit(X,y)
y_pred = logreg.predict(X)
print(metrics.accuracy_score(y,y_pred))
```
