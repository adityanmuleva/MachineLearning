```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
plt.style.use("seaborn-whitegrid")
```

```python
plt.subplot(2,2,1)
plt.plot([x for x in range(25)],[y**2 for y in range (25)],'-o')
plt.xlabel("Frequency")
plt.ylabel("Time")
plt.title("Test plot")
plt.show()


plt.subplot(2,2,2)
plt.bar([x for x in range(25)],[y**2 for y in range (25)])
plt.title("Test plot")
plt.show()

plt.subplot(2,2,3)
plt.hist([x for x in range(25)],[y**2 for y in range (25)])
plt.title("Test plot")
plt.show()

plt.subplot(2,2,4)
plt.scatter([x for x in range(25)],[y**2 for y in range (25)])
plt.title("Test plot")
plt.show()
```

```python
# Pie chart

names = ["aaditya", "arpita", "krishna", "jay", "gaurav"]
sizes = [5,4,3,2,1]
explode = [0.1,0,0,0,0]
hcm,ax1 = plt.subplots()
print(hcm)
ax1.pie(sizes, labels=names, explode=explode, shadow=True, startangle=90)
ax1.axis("equal")
plt.show()
```

```python
# customizing graphs
#1 -> 'ro' means colour-r,g,b,y and shape-o,^,-,p,h,t,s,d

plt.subplot(2,2,1)
plt.plot([x for x in range(25)],[y**2 for y in range (25)],'rd')
plt.xlabel("Frequency")
plt.ylabel("Time")
plt.title("Test plot")
plt.show()

plt.subplot(2,2,1)
m = np.arange(1,10,2)
plt.plot(m, m, 'r--', m,m**2,'bs',m,m**3,'gs')
plt.xlabel("Frequency")
plt.ylabel("Time")
plt.title("Test plot")
plt.show()
```

```python

a = np.sin([g**3 for g in range (100) ])
plt.subplot(2,2,2)
plt.hist(a)
plt.show()

plt.subplot(2,2,1)
plt.plot(a)
plt.show()

```

```python
f = np.arange(1,10,2)
t = np.arange(10,100,20)
d = np.sin(2*np.pi*f*t)*200
plt.plot(d,"r-o")
plt.show()
```

```python
x = np.linspace(start=-10, stop=10, num=101)
plt.plot(x, np.absolute(x), 'r-o')
plt.show()
```

```python
xx = x+ 1j *x[:, np.newaxis]
plt.imshow(np.abs(xx), extent=[-10,10,-10,10], cmap='hot')
plt.show()

xx = x+ 4j *x[:, np.newaxis]
plt.imshow(np.abs(xx), extent=[-10,10,-10,10], cmap='Blues')
plt.show()
```
